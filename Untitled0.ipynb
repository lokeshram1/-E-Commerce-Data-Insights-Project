```python
# ecommerce_data_insights/scripts/data_cleaning.py

import pandas as pd

def clean_data(input_path, output_path):
    # Load raw data
    df = pd.read_csv(input_path)

    # Drop duplicates
    df.drop_duplicates(inplace=True)

    # Convert OrderDate to datetime
    df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

    # Fill missing numeric values with median
    for col in ['Quantity', 'PricePerUnit', 'TotalPrice']:
        df[col].fillna(df[col].median(), inplace=True)

    # Fill missing categorical values with mode
    for col in ['Product', 'Category', 'PaymentMethod', 'City', 'Country']:
        df[col].fillna(df[col].mode()[0], inplace=True)

    # Recalculate TotalPrice
    df['TotalPrice'] = (df['Quantity'] * df['PricePerUnit']).round(2)

    # Save cleaned data
    df.to_csv(output_path, index=False)
    print(f"Cleaned data saved to {output_path}")


if __name__ == "__main__":
    input_file = "ecommerce_raw_data.csv"  # Adjusted for Colab
    output_file = "ecommerce_cleaned_data.csv"
    clean_data(input_file, output_file)


# ecommerce_data_insights/scripts/data_analysis.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("ecommerce_cleaned_data.csv")

# Set style
sns.set(style="whitegrid")

# Revenue by Product
plt.figure(figsize=(10, 6))
revenue_by_product = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)
revenue_by_product.plot(kind="bar", color="skyblue")
plt.title("Total Revenue by Product")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("revenue_by_product.png")
plt.show()

# Orders by Payment Method
plt.figure(figsize=(8, 5))
df["PaymentMethod"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Order Distribution by Payment Method")
plt.ylabel("")
plt.tight_layout()
plt.savefig("payment_method_distribution.png")
plt.show()

# Monthly Sales Trend
df['Month'] = pd.to_datetime(df['OrderDate']).dt.to_period('M')
monthly_sales = df.groupby('Month')["TotalPrice"].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(10, 6), color='green')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.tight_layout()
plt.savefig("monthly_sales_trend.png")
plt.show()
```

```python
# ecommerce_data_insights/scripts/data_cleaning.py

import pandas as pd

def clean_data(input_path, output_path):
    df = pd.read_csv(input_path)
    df.drop_duplicates(inplace=True)
    df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

    for col in ['Quantity', 'PricePerUnit', 'TotalPrice']:
        df[col].fillna(df[col].median(), inplace=True)

    for col in ['Product', 'Category', 'PaymentMethod', 'City', 'Country']:
        df[col].fillna(df[col].mode()[0], inplace=True)

    df['TotalPrice'] = (df['Quantity'] * df['PricePerUnit']).round(2)
    df.to_csv(output_path, index=False)
    print(f" Cleaned data saved to {output_path}")


if __name__ == "__main__":
    input_file = "ecommerce_raw_data.csv"
    output_file = "ecommerce_cleaned_data.csv"
    clean_data(input_file, output_file)


# ecommerce_data_insights/scripts/data_analysis.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load data
df = pd.read_csv("ecommerce_cleaned_data.csv")
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.to_period('M').astype(str)

# Seaborn style for static plots
sns.set(style="whitegrid")

# 1. Total Revenue by Product (Interactive)
revenue_product = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False).reset_index()
fig1 = px.bar(revenue_product, x="Product", y="TotalPrice", title=" Total Revenue by Product", color="TotalPrice",
              color_continuous_scale="Viridis", text_auto=True)
fig1.write_html("revenue_by_product.html")
fig1.show()

# 2. Order Distribution by Payment Method (Interactive Pie)
payment_counts = df["PaymentMethod"].value_counts().reset_index()
payment_counts.columns = ["PaymentMethod", "Count"]
fig2 = px.pie(payment_counts, names='PaymentMethod', values='Count', title=' Order Distribution by Payment Method',
              color_discrete_sequence=px.colors.qualitative.Pastel)
fig2.write_html("payment_method_distribution.html")
fig2.show()

# 3. Monthly Revenue Trend (Interactive Line Plot)
monthly_sales = df.groupby("Month")["TotalPrice"].sum().reset_index()
fig3 = px.line(monthly_sales, x='Month', y='TotalPrice', markers=True, title=' Monthly Sales Trend',
               line_shape='spline', color_discrete_sequence=['green'])
fig3.update_traces(marker=dict(size=8))
fig3.write_html("monthly_sales_trend.html")
fig3.show()

# 4. Top Cities by Revenue
top_cities = df.groupby("City")["TotalPrice"].sum().sort_values(ascending=False).head(10).reset_index()
fig4 = px.bar(top_cities, x="City", y="TotalPrice", title=" Top 10 Cities by Revenue", color="TotalPrice",
              text_auto=True, color_continuous_scale="Blues")
fig4.write_html("top_cities_by_revenue.html")
fig4.show()

# 5. Heatmap: Revenue by Category and Month
pivot = df.pivot_table(index='Category', columns='Month', values='TotalPrice', aggfunc='sum').fillna(0)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title(" Revenue Heatmap by Category and Month")
plt.tight_layout()
plt.savefig("revenue_heatmap.png")
plt.show()
```
